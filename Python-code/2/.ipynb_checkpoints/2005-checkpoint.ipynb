```python
# 内置函数
```

```python
# abs 绝对值

print(abs(-1))
print(abs(-1.1))
print(abs())
print(abs('-1.1'))
```

```python
# all 均为真值 返回 true     --- 有点像 js 的 every

print(all([True, 1, "hello", [1, 2]])) 

# any 任意元素为真值返回 true     --- js 的 some

print(any([False, 0, 'python']))

# bool 
print(bool(0))
print(bool([]))

# filter     --- 类似于 js 中的 filter
result = filter(lambda x: x > 3, [1, 2, 3, 4, 5])
print(result) 
print(list(result))

# len  字符串（str）、列表（list）、元组（tuple）、字典（dict）、集合（set）、范围（range）、等其他可迭代对象
print(len('python'))

print(max([1,2,3,4]))
print(min([1,2,3,4]))

# pow  次幂，第三个参数 取余 2 的 3 次幂 % 5
print(pow(2,3))
print(pow(2,3,5))

# reversed
print(list(reversed([1,2,3,4,5])))

# round     --- 类似 js 的 round
print(round(10.45))

# sum
print(sum([1,2,3,4,5]))
print(sum([1,2,3,4,5], 2))  # 第二个参数 起始值
```

```python
# bin 整数转换为带前缀 '0b' 的二进制字符串    -  整数

print(bin(10))
print(bin(-10))
print(bin(10.2))
```

```python
# eval
x = 2
print(eval('x * 3'))
```
