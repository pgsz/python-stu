```python
456 > 1234  # False
```

```python
'456' > '1234' # True  字符串比较不是比较数值大小，而是按字符逐个比较：逐个字符比较 Unicode 编码值
'2' > '10'  # True
```

```python
len('456') > len('1234')  # False
```
