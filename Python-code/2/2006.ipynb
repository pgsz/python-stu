```python
# 布尔运算

# 类似于 js 的 || 
print('12' or '22')
print('0' or '22')

# 类似于 js 的 &&
print('12' and '22')
print('0' and '22')

#
print(not 12 == 12)
```

```python
# 比较运算
print(2 >= 1)
```

```python
# 数字类型
print(int('12'))
# 商 和 余数
print(divmod(22, 5))
print(10 / 4)
```

```python
# 序列类型  list、tuple、range
# list 可以动态修改，tuple 是不可变的（但若元素是可变对象，如列表，其内部可修改），range 生成固定序列，不可修改

# list：方括号
list_obj = [1, 2, 3]
print(list_obj)

# tuple：圆括号（逗号是关键）
tuple_obj = (1, 2, 3)
print(tuple_obj)

# range：range() 函数
range_obj = range(1, 4)  # 生成 1, 2, 3
print(range_obj)

print(1 in [1,2,3,4])
print(1 not in [1,2,3,4])

lists = [1] * 3
print(lists)
lists[0] = 3
print(lists)

# 注意 与 range 的区别
lists1 = [[]] * 3
print(lists1)
lists1[0].append(3)
print(lists1) # [[3], [3], [3]]

lists2 = [[] for i in range(3)]
print(lists2)
lists2[0].append(3)
print(lists2) # [[3], [], []]

# 可以直接相加 拼接
print([1,2,3] + [4,5,6]) # [1, 2, 3, 4, 5, 6]
```
